```python
""" Problem(2 and 3)
Created on Monday May 7 22:05:02 2023

@author: Sayan Sircar
"""
""" Compressibility vs P plot for three gases Problem 2 third part"""

from scipy.optimize import fsolve
import numpy as np
import matplotlib.pyplot as plt
R = 0.08206
#Pc =a/27*b**2
#Tc =8*a/27*R*b
Pcn =33.69
Tcn =126.45
a = 27 * R**2 * Tcn**2 / (Pcn * 64)
b = R * Tcn / (8 * Pcn)
print(a)
print(b)
TrN =2.372
def function1(V, Pr):
    Pn = Pr * Pcn
    Tn = TrN * Tcn
    return Pn* (V - b) - (R * Tn)  +  a / V**2 * (V - b)
Prsol1 = np.linspace(0.1,50)
Prsol2 = np.linspace(100,1000)
V1a = [fsolve(function1, 3, args=(Pr,))[0] for Pr in Prsol1]
V1b = [fsolve(function1, 3, args=(Pr,))[0] for Pr in Prsol2]
Tn = TrN * Tcn
Pna = Prsol1 * Pcn
Zna = Pna * V1a / (R * Tn)
Pnb = Prsol2 * Pcn
Znb = Pnb * V1b / (R * Tn)
Pcm=45.52
Tcm=190.82
a1 = 27 * R**2 * Tcm**2 / (Pcm * 64)
b1 = R * Tcm / (8 * Pcm)
print(a1)
print(b1)
TrM=1.5721
def function2(V, Pr):
    Pm = Pr * Pcm
    Tm = TrM * Tcm
    return Pm * (V - b1) - (R * Tm)  +  a1 / V**2 * (V - b1)
Prsol1 = np.linspace(0.1,50)
Prsol2 = np.linspace(100,1000)
V2a = [fsolve(function2, 3, args=(Pr,))[0] for Pr in Prsol1]
V2b = [fsolve(function2, 3, args=(Pr,))[0] for Pr in Prsol2]
Tm = TrM * Tcm
Pma = Prsol1 * Pcm
Zma = Pma * V2a / (R * Tm)
Pmb = Prsol2 * Pcm
Zmb = Pmb * V2b / (R * Tm)
PcHe=2.242823
TcHe=11.51
a2 = 27 * R**2 * TcHe**2 / (PcHe * 64)
b2 = R * TcHe / (8 * PcHe)
print(a2)
print(b2)
TrHe=26.06429
def function3(V, Pr):
    PHe = Pr * PcHe
    THe = TrHe * TcHe
    return PHe * (V - b2) - (R * THe)  +  a2 / V**2 * (V - b2)
Prsol1 = np.linspace(0.1,50)
Prsol2=np.linspace(100,1000)
V3a = [fsolve(function3, 3, args=(Pr,))[0] for Pr in Prsol1]
V3b=  [fsolve(function3, 3, args=(Pr,))[0] for Pr in Prsol2]
THe = TrHe * TcHe
PHea = Prsol1 * PcHe
PHeb=Prsol2*PcHe
ZHea = PHea * V3a / (R * THe)
ZHeb = PHeb * V3b / (R * THe)
fig=plt.figure()
plt.plot(Prsol1, Zna,label='For nitogen(a=1.3483,b=0.038577')
plt.plot(Prsol1, Zma,label='For methane(a=2.2725,b=0.043067')
plt.plot(Prsol1, ZHea,label='For Helium(a=0.034145,b=0.023733)')
plt.xlabel('$P$ in units of bar(1bar=0.987atm)')
plt.ylabel('Z(unitless considering 1 mole quantity of each gases)')
plt.legend()
plt.grid(color='black')
plt.title('Compressibility factor vs P curves for nitrogen,methane and Helium in medium low presure range at T=300K')
plt.savefig("Compressibility factor plot.pdf")
plt.xlim([0,50])
plt.ylim([0,3])
fig=plt.figure()
plt.plot(Prsol2, Znb,label='For nitogen(a=1.3483,b=0.038577')
plt.plot(Prsol2, Zmb,label='For methane(a=2.2725,b=0.043067')
plt.plot(Prsol2, ZHeb,label='For Helium(a=0.034145,b=0.023733)')
plt.xlabel('$P$ in units of bar(1bar=0.987atm)')
plt.ylabel('Z(unitless considering 1 mole quantity of each gases)')
plt.legend()
plt.grid(color='black')
plt.title('Compressibility factor vs P curves for nitrogen,methane and Helium in high pressure at T=300K')
plt.savefig("Compressibility factor plot 2.pdf")
plt.xlim([100,1000])
plt.ylim([0,80])
```
